### Enunciado:
En este primer trabajo comenzaremos por diseñar un generador de señales que utilizaremos en las primeras 
simulaciones que hagamos. La primer tarea consistirá en programar una función que genere señales senoidales
y que permita parametrizar:

- La amplitud máxima de la senoidal (volts)
- Su valor medio (volts)
- La frecuencia (Hz)
- La fase (radianes)
- La cantidad de muestras digitalizada por el ADC (# muestras)
- La frecuencia de muestreo del ADC.

Es decir que la función que ustedes armen debería admitir se llamada de la siguiente manera

tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)
Recuerden que tanto xx como tt deben ser vectores de Nx1. Puede resultarte útil el módulo de visualización 
matplotlib.pyplot donde encontrarán todas las funciones de visualización estilo Matlab. Para usarlo:

import matplotlib.pyplot as plt
plt.plot(tt, xx)

Implemento las librerias a usar. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defino mi funcion seno. 

```python
def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
    tt=np.arange(0,nn,1)*(1/fs)
    xx=dc + vmax*np.sin(2*np.pi*ff*tt + ph)
    return tt,xx 
```

Se llama a la función pasándole como parámetros la amplitud máxima de la señal senoidal, el valor medio, la frecuencia, la fase, el número de muestras tomadas por el ADC y la frecuencia de muestreo del ADC.

```python
tt, xx = mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph=0, nn=100, fs = 100)
```

Se grafica la función a partir de los parámetros obtenidos anteriormente. 

```python
plt.plot(tt, xx)
plt.title("Funcion Senoidal")
plt.xlabel("Tiempo [t]")
plt.ylabel("Voltaje [V]")
plt.grid(ls="--")
plt.legend()
plt.show()
```
